```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Replace 'your_data.csv' with the actual file path
data = pd.read_csv('marginal.csv')
```

```python
data = {
    'Age': [20, 30, 40, None, 30, 40, 20, 30, 40],  # Adding a missing value
    'Industrial_Category': ['N to O - Females', 'P to Q - Persons', 'P to Q - Males',
                            'P to Q - Females', 'R to U - HHI - Persons', 'R to U - HHI - Males',
                            'R to U - HHI - Females', 'R to U - Non HHI - Persons', 'R to U - Non HHI - Males'],
    'Sex': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male'],
    'Count': [100, 200, 120, 80, 150, 60, 90, 130, 50]
}
df = pd.DataFrame(data)

# Handling missing values
df = df.dropna()

# Print the cleaned data
print(df)
```

```python
# Data aggregation
agg_data = df.groupby(['Age', 'Industrial_Category', 'Sex']).size().reset_index(name='Count')

# Print the aggregated data
print(agg_data)
```

```python
# Plot Age Distribution
plt.figure(figsize=(12, 6))
sns.barplot(x='Age', y='Count', data=agg_data, hue='Sex')
plt.title('Age Distribution of Marginal Workers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(title='Sex', loc='upper right')
plt.show()
```

```python
data = {
    'Industrial Category': ['N to O - Females', 'P to Q - Persons', 'P to Q - Males', 'P to Q - Females',
                            'R to U - HHI - Persons', 'R to U - HHI - Males', 'R to U - HHI - Females',
                            'R to U - Non HHI - Persons', 'R to U - Non HHI - Males', 'R to U - Non HHI - Females'],
    'Count': [100, 200, 120, 80, 150, 60, 90, 130, 50, 80]
}

df = pd.DataFrame(data)

# Extracting category information for better plotting
df['Category'] = df['Industrial Category'].str.split('-').str[0].str.strip()

# Plotting
plt.figure(figsize=(12, 8))
sns.barplot(x='Category', y='Count', hue='Industrial Category', data=df)
plt.title('Distribution of Marginal Workers in Industrial Categories')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
# Sex Distribution Plot
plt.figure(figsize=(8, 6))
sns.barplot(x='Sex', y='Count', data=agg_data)
plt.title('Sex Distribution of Marginal Workers')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()
```

```python
# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['Count'], c=df['Sex'].map({'Male': 'b', 'Female': 'r'}), label=df['Sex'], marker='o')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Scatter Plot')
plt.legend()
plt.show()
```
